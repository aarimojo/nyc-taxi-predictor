# Reading de data from May 2022

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns

data_folder_path = "../data/raw/yellow_tripdata_2022-05.parquet"
df = pd.read_parquet(data_folder_path)
print(f"Df shape with NaN{df.shape}")
df = df.dropna()
pd.set_option('display.float_format', lambda x: '%.3f' % x)
print(f"Df shape without NaN{df.shape}")
df.head(25)
```

```python
#Remove columns of date and letters
df_modified=df.drop(["tpep_pickup_datetime","tpep_dropoff_datetime","store_and_fwd_flag","VendorID","RatecodeID","PULocationID","DOLocationID"], axis="columns")
df_modified
```

```python
correlation_matrix = df_modified.corr()

# Create the heatmap
plt.figure(figsize = (10,8))
sns.heatmap(correlation_matrix, cmap = 'coolwarm')
plt.show()
```

```python
#Search for null values
print(df_modified.isnull().sum())
```

```python
df_modified['trip_distance'].hist(bins=20)
plt.show()

# Boxplot
sns.boxplot(data=df_modified, x='trip_distance')
plt.show()
```
